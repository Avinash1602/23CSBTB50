```python
l1=[1,23,45,67]
l2=[34,49,23,64]
print(l1+l2)
```

```python
import numpy as np
a=np.array(l1)
b=np.array(l2)
print("Add = ",a+b)
print("Mul = ",a*b)
print("Sub = ",a-b)
print("Div = ",a/b)
```

```python
print("Mean of 'a' = ",np.mean(a))
print("Variance of 'a' = ",np.var(a))
print("Minimum value of 'a'  ",np.min(a))
print("Maximum value of 'a' = ",np.max(a))
print("Standard Deviation of 'a' = ",np.var(a)**0.5)
```

```python
A=np.array([[2,5,1],[7,3,4],[9,5,0]])
B=np.array([[1,8,3],[2,6,7],[8,3,4]])
print("Matrix A : \n",A)
print("\n")
print("Matrix B : \n",B)
```

```python
#Transpose of a Matrix :
a=np.transpose(A)
print("Transpose Matrix of A : \n",a)
b=np.transpose(B)
print("\n")
print("Transpose Matrix of B : \n",b)
```

```python
#Trace of a Matrix :
T1=np.trace(a)
print("Trace of Transpose Matrix 'a' : ",T1)
T2=np.trace(b)
print("Trace of Transpose Matrix 'b' : ",T2)
```

```python
import matplotlib.pyplot as plt
d={'population':[100,200,300,400,500],
    'households':[20,87,126,189,236]}
x=d['population']
y=d['households']
plt.plot(x,y,'purple')
plt.scatter(x,y,c='black')
plt.xlabel("Population")
plt.ylabel("Households")
plt.title("Population vs. Households")
plt.grid(True,c='red')
```
